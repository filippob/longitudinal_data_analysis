```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Create synthetic data
np.random.seed(0)
X = np.sort(np.random.rand(30) * 10)
y = np.sin(X) + np.random.normal(0, 0.2, size=X.shape)  # underlying function is sinusoidal

# Plot settings
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Data points')

# Degrees to try
degrees = [1, 2, 5, 10]
colors = ['blue', 'green', 'red', 'purple']

# Plot each model
X_plot = np.linspace(0, 10, 100).reshape(-1, 1)  # smoother curve for plotting

for degree, color in zip(degrees, colors):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X.reshape(-1, 1), y)
    y_plot = model.predict(X_plot)
    plt.plot(X_plot, y_plot, label=f'Degree {degree}', color=color)

plt.legend()
plt.title('Polynomial Regression with Different Degrees')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
